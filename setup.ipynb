# Utility file

```python
input_path = '../classified_images'
output_path = '../train_test_images'
seed = 42
```

1. Converts images to .jpg

```python
import os
from PIL import Image
```

```python
# Loop through subfolders and convert images
for folder in os.listdir(input_path):
    if '.DS_Store' in folder:
        continue
    
    folder_path = os.path.join(input_path, folder)
    
    for file in os.listdir(folder_path):
        if file.endswith('.webp'):
            img_path = os.path.join(folder_path, file)
            img = Image.open(img_path).convert('RGB')
            
            # Save the image as .jpg in the same folder
            img.save(os.path.join(folder_path, file.replace('.webp', '.jpg')))
            
            # Remove the original .webp file after conversion
            os.remove(img_path)
```

2. Splits images into train/val directories

```python
import splitfolders
from PIL import Image
import os
```

```python
splitfolders.ratio(input_path, output=output_path, seed=seed, ratio=(.8, .2))
```
