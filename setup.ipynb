# Utility file

```python
import os
from PIL import Image
from tqdm import tqdm
```

```python
input_path = '../camera_frames'
output_path = '../train_test_images'
seed = 42
```

1. Converts images to .jpg

```python
# Loop through subfolders and convert images    
for file in tqdm(os.listdir(input_path)):
    if file.endswith('.webp'):
        img_path = os.path.join(input_path, file)
        img = Image.open(img_path).convert('RGB')
        
        # Save the image as .jpg in the same folder
        img.save(os.path.join(input_path, file.replace('.webp', '.jpg')))
        
        # Remove the original .webp file after conversion
        os.remove(img_path)
```
